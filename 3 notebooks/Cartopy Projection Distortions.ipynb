# Showcasing cartopy projection distortions

Here, I show multiple views of Illinois with different map projections. Some of these take a while to run because cartopy has to convert the PlateCarree extent coordinates into the respective projection's coordinates.

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np

latN = 43
latS = 36
lonE = -85
lonW = -95



a = plt.figure(figsize=(11, 8.5))

fig = plt.axes(projection=ccrs.PlateCarree())
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)

```

```python
a = plt.figure(figsize=(11, 8.5))

fig = plt.axes(projection=ccrs.Mollweide(central_longitude= -88))
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)
```

```python
a = plt.figure(figsize=(11, 8.5))

fig = plt.axes(projection=ccrs.AlbersEqualArea(central_longitude= -88))
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)
```

```python
plt.figure(figsize=(11, 8.5))

fig = plt.subplot(1, 1, 1, projection = ccrs.EckertII())
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)
```

```python
plt.figure(figsize=(11, 8.5))

fig = plt.subplot(1, 1, 1, projection = ccrs.EckertII(central_longitude= -88))
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)
```
