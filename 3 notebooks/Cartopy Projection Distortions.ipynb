# Showcasing cartopy projection distortions

Here, I show multiple views of Illinois with different map projections. Some of these take a while to run because cartopy has to convert the PlateCarree extent coordinates into the respective projection's coordinates. A line is also plotted across the map to show how it is transformed based off of the distortion. When adding features like lines, contours, or points, the code may take longer to run depending on the projection used as the coordinates need to be transformed from Plate Carree, which has equal space between every latitude/longitude line, to the used projection. 

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np

latN = 43
latS = 36
lonE = -85
lonW = -95



a = plt.figure(figsize=(11, 8.5))

fig = plt.axes(projection=ccrs.PlateCarree())
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)

fig.plot([-90, -89], [40, 42], transform= ccrs.PlateCarree())


plt.show()
```

```python
a = plt.figure(figsize=(11, 8.5))

fig = plt.axes(projection=ccrs.Mollweide(central_longitude= -88))
fig.set_extent([lonW, lonE, latS, latN], crs = ccrs.PlateCarree())
fig.coastlines()
fig.add_feature(cfeature.BORDERS)
fig.add_feature(cfeature.STATES)
fig.add_feature(cfeature.LAND)


fig.plot([-90, -89], [40, 42], transform= ccrs.PlateCarree())


plt.show()
```
